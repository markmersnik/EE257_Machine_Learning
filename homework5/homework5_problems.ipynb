```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#Problem 3 on page 332
p = np.stack(i*0.01 for i in range(100))

gini = []
class_error = []
entropy = []

for i in p:
    gini.append(2*i*(1-i))
    class_error.append(1-max(i, 1-i))
    entropy.append((-i*np.log(i)-(1-i)*np.log(1-i)))
plt.figure(figsize=(16,12))
plt.scatter(p, gini, color='green')
plt.scatter(p, class_error, color='red')
plt.scatter(p, entropy, color='blue')
plt.show()
```
